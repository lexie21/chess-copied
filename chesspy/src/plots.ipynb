```python
from matplotlib.pyplot import subplots
```

```python
import numpy as np
rng = np.random.default_rng(5)
x = rng.standard_normal(size=50)
y = rng.normal(0,5,50)
```

```python
print(np.mean(x))
print(np.std(x))
print(np.mean(y))
print(np.std(y))
```

```python
fig, ax = subplots(figsize=(8,8))
ax.scatter(x,y)
```
